```python
df["Ticket_Price_Cr"] = (
    df["Ticket_Price_Cr"]
    .astype(str)
    .str.replace(r"[^\d.]", "", regex=True)   # remove non-numeric chars
    .replace("", np.nan)                      # convert blanks to NaN
    .astype(float)
)

# Fill NaN with mean (safe assignment, no warning)
mean_price = df["Ticket_Price_Cr"].mean()
df["Ticket_Price_Cr"] = df["Ticket_Price_Cr"].fillna(mean_price)
```

```python
df["Amenity_Score"]=df["Amenity_Score"].fillna(df["Amenity_Score"].mean())
df["Unit_Size_Sqft"]=df["Unit_Size_Sqft"].fillna(df["Unit_Size_Sqft"].mean())
df['Buyer_Comments']=df['Buyer_Comments'].fillna('No Coments')
```

```python
df["Price_per_Sqft"] = (df["Ticket_Price_Cr"] * 10000000) / df["Unit_Size_Sqft"]
```

```python
df["Booking_Flag"] = df["Transaction_Type"].apply(
        lambda x: 1 if str(x).lower() == "primary" else 0)
```

```python
df['Purchase_Quarter'] = pd.to_datetime(df['Purchase_Quarter'])
df['Quarter_Number'] = df['Purchase_Quarter'].dt.quarter
```

```python
df=df.drop_duplicates()
```
